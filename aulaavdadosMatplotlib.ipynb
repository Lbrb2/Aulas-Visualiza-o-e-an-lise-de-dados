```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ["NVIDIA","Apple","Alphabet","Microsoft","Amazon","Saudi Aramco","TSMC","Broadcom","Meta Platforms","Tesla"]
```

```python
trilhoes = [4.311,3.761,3.562,2.775,2.251,1.778,1.757,1.491,1.453,1.353]
```

```python
plt.plot(empresas,trilhoes)
```

```python
plt.plot(empresas,trilhoes)
plt.title("Empresas mais ricas do mundo")
```

```python
plt.plot(empresas,trilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Trilhoes")
```

```python
plt.plot(empresas,trilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Trilhoes")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,8)
plt.plot(empresas,trilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Trilhões")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,8)
plt.plot(empresas,trilhoes)
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas")
plt.ylabel("Trilhões")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,8)
plt.plot(empresas,trilhoes)
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas",fontsize=20)
plt.ylabel("Trilhões",fontsize=20)
plt.show()
```

```python
plt.plot(empresas,trilhoes, color = "purple")
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas",fontsize=20)
plt.ylabel("Trilhões",fontsize=20)
plt.legend(["Trilhões de Dólares"], loc = "right",fontsize=15)
```

```python
plt.plot(empresas,trilhoes, color = "purple",lw=10)
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas",fontsize=20)
plt.ylabel("Trilhões",fontsize=20)
plt.legend(["Trilhões de Dólares"], loc = "upper right",fontsize=15)
```

```python
plt.plot(empresas,trilhoes, color = "purple",marker = "o")
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas",fontsize=20)
plt.ylabel("Trilhões",fontsize=20)
plt.legend(["Trilhões de Dólares"], loc = "right",fontsize=15)
```

```python
plt.plot(empresas,trilhoes, color = "purple",marker = "o")
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas",fontsize=20)
plt.ylabel("Trilhões",fontsize=20)
plt.legend(["Trilhões de Dólares"], loc = "right",fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas,trilhoes, color = "purple")
plt.title("Empresas mais ricas do mundo",fontsize=25)
plt.xlabel("Empresas",fontsize=20)
plt.ylabel("Trilhões",fontsize=20)
plt.legend(["Trilhões de Dólares"], loc = "center right",fontsize=15)
plt.show()
```

```python
cores = ["black","dimgray","dimgrey","gray","grey","darkgray","darkgrey","silver","lightgray","lightgrey"]
```

```python
plt.bar(empresas,trilhoes,color = cores,edgecolor = "black")
plt.title("Empresas mais ricas do mundo",fontsize = 25)
plt.xlabel("Empresas",fontsize = 20)
plt.ylabel("Trilhões",fontsize = 20)
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8-white")
plt.bar(empresas,trilhoes,color = cores,edgecolor = "black")
plt.title("Empresas mais ricas do mundo",fontsize = 25)
plt.xlabel("Empresas",fontsize = 20)
plt.ylabel("Trilhões",fontsize = 20)
plt.show()
```

```python
plt.style.use("seaborn-v0_8-white")
plt.barh(empresas,trilhoes,color = cores,edgecolor = "black")
plt.title("Empresas mais ricas do mundo",fontsize = 25)
plt.xlabel("Empresas",fontsize = 20)
plt.ylabel("Trilhões",fontsize = 20)
plt.show()
```
