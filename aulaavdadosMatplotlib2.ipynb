Podemos selecionar a coluna diretamente:

```python
dados.Age.unique()
```

```python
np.sort(dados.Age.unique())
```

Para usarmos a seleção direta de colunas com o ponto,precisamos ter certeza de que as colunas não contêm espaços ou caracteres especiais permitidos nas palavras entre as aspas.

```python
df = dados.copy()
df.columns = [column.replace(" ","_")for column in df.columns]
df.columns
```

```python
idade = dados.Age

plt.hist(idade, label="Idades",color = "purple")
plt.title("Distribuição de Idades")
plt.xlabel("Idades")
plt.ylabel("Quantidade")
plt.legend()
plt.show()
```
